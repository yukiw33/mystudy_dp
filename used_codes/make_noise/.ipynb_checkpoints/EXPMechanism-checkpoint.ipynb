## 時刻tに対してノイズをかけた結果の可視化

```python
# モジュールの導入
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from scipy.special import lambertw as LW
```

### 付与する効用関数を可視化していく

#### 効用関数10:1

```python
epsilon = 20
```

```python
t__ = np.linspace(-500, 1000, 1000)
U__ = []
for i in range(len(t__)):
    if t__[i] < 0:
        U__.append(10*t__[i])
    else:
        U__.append(-t__[i])

fig, ax = plt.subplots()
ax.plot(t__, U__)
ax.set_xlabel("t (h)")
ax.set_ylabel("U (utility function)")
plt.title("utility function 10:1")
plt.minorticks_on()
plt.grid()
#plt.savefig("utility_graph.png", dpi=120)
plt.show()
```

これをpdf,cdfとして可視化していく．

```python
# pdfを可視化する
t = np.linspace(-10, 20, 1000)
U = []
k = 1
for i in range(len(t)):
    if t[i] < 0:
        tmp = ( epsilon / 2*(k+1) ) * np.e**( (epsilon*t[i])/2 )
        U.append(tmp)
    else:
        tmp = ( epsilon / 2*(k+1) ) * np.e**( -(epsilon*t[i])/(2*k) )
        U.append(tmp)
fig, ax = plt.subplots()
ax.plot(t, U)
ax.set_xlabel("t (h)")
ax.set_ylabel("pdf (probability density function)")
plt.title("probability density function 10:1")
plt.minorticks_on()
plt.grid()
#plt.savefig("pdf_graph20__1.png", dpi=120)
plt.show()
```

```python
# cdfを可視化する
CDF = []
for i in range(len(t)):
    if t[i] < 0:
        tmp = ( 1 / (k+1) ) * np.e**( (epsilon * t[i]) / 2)
        CDF.append(tmp)
    else:
        tmp = 1 - ( k / (k+1) ) * np.e**((-epsilon * t[i]) / (2*k))
        CDF.append(tmp)
fig, ax = plt.subplots()
ax.plot(t, CDF)
ax.set_xlabel("t (h)")
ax.set_ylabel("cdf (cumulative distribution function)")
plt.title("cumulative distribution function 10:1")
plt.minorticks_on()
plt.grid()
# plt.savefig("cdf_graph20.png", dpi=120)
plt.show()
```

```python
# cdf^(-1)の可視化
CDF_1 = []
y_ = []
y = np.linspace(0,1,1000+2)
for i in range(len(y)):
    y_.append(y[i])
y_.pop(0)
y_.pop(-1)
for i in range(len(y_)):
    if y_[i] < (1 / (k+1)):
        CDF_1.append((2/epsilon)*np.log((k+1)*y_[i]))
    else:
        CDF_1.append(((2*k)/epsilon)*np.log(k/((k+1)*(1-y_[i]))))
fig, ax = plt.subplots()
ax.plot(y_, CDF_1)
ax.set_xlabel("random number 0~1")
ax.set_ylabel("cdf inv (cumulative distribution function inverse)")
plt.title("cumulative distribution function inverse 10:1, epsilon=0.2")
plt.minorticks_on()
plt.grid()
# plt.savefig("cdf_inverse_graph_02.png", dpi=120)
plt.show()
```

#### この分布にしたがってノイズ付与されているか確認する

```python
p_t = np.random.uniform(0, 1, 10000)
t_1 = []
for i in range(len(p_t)):
    if p_t[i] <= (1 / (k+1)):
        t_1.append((2 / (epsilon*4)) * np.log( (k+1) * p_t[i]))
    else:
        t_1.append(((2*k) / (epsilon*4)) * np.log( k / ((k+1)*(1-p_t[i])) ))
```

```python
t_2 = np.round(t_1, decimals=2)
```

```python
plt.hist(t_2, bins=1000)
```

#### 3次元描画していく

```python
# ランダム化された点の分布を見るための数の定義
Num = 100000

# ノイズを付与した座標の値を格納するためのリストの定義
zx_list = []
zy_list = []

# 実際の位置の定義
real_x = 5.0
real_y = 3.0

# 実際の時間を定義(今の時間を0と定義してノイズをかける)
real_t = 0.0

# ノイズを付与した時刻の値を格納するためのリストの定義
zt_list = []
```

```python
theta = np.random.uniform(0, 2*np.pi, Num)

p = np.random.uniform(0, 1, Num)
W = LW( ((p-1) / np.e), k = -1)
r = ( -1 / epsilon ) * (W + 1)
z_x = real_x + r * np.cos(theta)
z_y = real_y + r * np.sin(theta)

zx_list.append(np.array(z_x))
zy_list.append(np.array(z_y))

p_t = np.random.uniform(0, 1, Num)
for i in range(Num):
    if p_t[i] <= (1 / (k+1)):
        t_ = (2 / (epsilon*4)) * np.log( (k+1) * p_t[i])
    else:
        t_ = ((2*k) / (epsilon*4)) * np.log( k / ((k+1)*(1-p_t[i])) )
    z_t = real_t + t_
    zt_list.append(np.array(z_t))
zx__ = []
zy__ = []
for i in range(Num):
    zx__.append(np.array(zx_list[0][i]))
    zy__.append(np.array(zy_list[0][i]))
```

```python
# 散布図としてノイズを付与した座標を出力する
fig, ax = plt.subplots()
plt.scatter(np.real(zx__), zt_list)
# 実際の位置を赤く表示
plt.scatter(real_x, real_t, color="red")
ax.set_xlabel("x (km)")
ax.set_ylabel("t' (s)")
plt.title("scatter X T , epsilon=20")
plt.grid()
plt.savefig("scatter_exp_xt.png", dpi=120)
plt.show()
```

```python
zx_ = np.round(zx__, decimals=2)
zt_ = np.round(zt_list, decimals=2)
# ヒストグラムを出力する
hist, xedges, yedges = np.histogram2d(zx_, zt_, bins=100, range=[[4, 6], [-1, 3.5]])

xpos, ypos = np.meshgrid(xedges[:-1] + 1, yedges[:-1] + 1, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

dx = dy = 0.01
dz = hist.ravel()

fig = plt.figure(figsize=(16,16))
ax = fig.add_subplot(111, projection="3d")
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')
ax.set_xlabel("x (km)")
ax.set_ylabel("t' (s)")
ax.set_zlabel("density")
ax.view_init(elev=40, azim=60)
plt.title("PL-EX Mechanism epsilon=20", y=-0.005)
plt.savefig("3d_histogram_exp_20.jpg", dpi=120)
plt.show()
```
