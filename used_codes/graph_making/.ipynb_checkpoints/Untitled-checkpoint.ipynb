```python
# モジュールの導入
import numpy as np
from scipy.special import lambertw as LW
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import statistics
import pandas as pd
```

```python
x_1 = np.linspace(-20,20,1024)
x_2 = np.linspace(5,40,1024)
f_1 = []
f_2 = []
b = 10
m_1 = 0
m_2 = 25

for i in range(len(x_1)):
    f_1.append((1/(2*b))*np.e**((-abs(x_1[i]-m_1))/b))
for i in range(len(x_2)):
    f_2.append((1/(2*b))*np.e**((-abs(x_2[i]-m_2))/b))
```

```python
#fig, ax = plt.subplots()
#ax.plot(x_1, f_1)
#ax.plot(x_2, f_2)
fig = plt.figure(figsize=(15,5))
plt.plot(x_1, f_1)
plt.plot(x_2, f_2)
plt.vlines([0], 0.0, 0.049, "black", linestyles='dashed')
plt.vlines([25], 0.0, 0.049, "black", linestyles='dashed')
plt.xticks([])
plt.yticks([])
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.savefig("pdf_1_graph.png")
plt.show()
```
